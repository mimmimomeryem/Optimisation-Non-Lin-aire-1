```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def eggholder(xx):
    x1 = xx[0]
    x2 = xx[1]

    term1 = -(x2 + 47) * np.sin(np.sqrt(np.abs(x2 + x1 / 2 + 47)))
    term2 = -x1 * np.sin(np.sqrt(np.abs(x1 - (x2 + 47))))

    y = term1 + term2
    return y

# Créer une grille de points (x, y)
x = np.linspace(-512, 512, 400)
y = np.linspace(-512, 512, 400)
X, Y = np.meshgrid(x, y)
Z = np.zeros_like(X)

# Calculer la valeur de la fonction pour chaque point de la grille
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        Z[i, j] = eggholder([X[i, j], Y[i, j]])

# Coordonnées du minimum global
x_min = 512
y_min = 404.2319
z_min = eggholder([x_min, y_min])

# Créer une grille de points (x, y) pour la vue restreinte
x_zoom = np.linspace(400, 600, 400)
y_zoom = np.linspace(512, 612, 400)
X_zoom, Y_zoom = np.meshgrid(x_zoom, y_zoom)
Z_zoom = np.zeros_like(X_zoom)

# Calculer la valeur de la fonction pour chaque point de la grille restreinte
for i in range(X_zoom.shape[0]):
    for j in range(X_zoom.shape[1]):
        Z_zoom[i, j] = eggholder([X_zoom[i, j], Y_zoom[i, j]])

# Créer la figure avec deux sous-graphiques
fig = plt.figure(figsize=(14, 7))

# Couleur de la carte
cmap = 'magma'

# Premier sous-graphe : vue d'ensemble
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, Z, cmap=cmap, alpha=0.8)

# Ajouter le minimum global
ax1.scatter(x_min, y_min, z_min, color='r', s=100)  # point rouge pour le minimum global
ax1.text(x_min, y_min, z_min, 'Minimum Global', color='red')

# Ajouter des labels et un titre
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
ax1.set_title('Fonction Porte-Oeufs (Eggholder) - Vue d\'ensemble')

# Deuxième sous-graphe : vue restreinte
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(X_zoom, Y_zoom, Z_zoom, cmap=cmap, alpha=0.8)

# Ajouter le minimum global
ax2.scatter(x_min, y_min, z_min, color='r', s=100)  # point rouge pour le minimum global
ax2.text(x_min, y_min, z_min, 'Minimum Global', color='red')

# Ajouter des labels et un titre
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')
ax2.set_title('Fonction Porte-Oeufs (Eggholder) - Zoom sur le minimum')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def cross_in_tray(xx):
    x = xx[0]
    y = xx[1]
    term1 = np.sin(x) * np.sin(y)
    term2 = np.exp(np.abs(100 - np.sqrt(x**2 + y**2) / np.pi))
    return -0.0001 * (np.abs(term1 * term2) + 1) ** 0.1

# Créer une grille de points (x, y) pour la vue d'ensemble
x = np.linspace(-10, 10, 400)
y = np.linspace(-10, 10, 400)
X, Y = np.meshgrid(x, y)
Z = np.zeros_like(X)

# Calculer la valeur de la fonction pour chaque point de la grille
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        Z[i, j] = cross_in_tray([X[i, j], Y[i, j]])

# Coordonnées des minima globaux
minima = [
    (1.3491, -1.3491),
    (-1.3491, 1.3491),
    (1.3491, 1.3491),
    (-1.3491, -1.3491)
]
z_min = cross_in_tray([minima[0][0], minima[0][1]])

# Créer une grille de points (x, y) pour la vue restreinte
x_zoom = np.linspace(-2, 2, 400)
y_zoom = np.linspace(-2, 2, 400)
X_zoom, Y_zoom = np.meshgrid(x_zoom, y_zoom)
Z_zoom = np.zeros_like(X_zoom)

# Calculer la valeur de la fonction pour chaque point de la grille restreinte
for i in range(X_zoom.shape[0]):
    for j in range(X_zoom.shape[1]):
        Z_zoom[i, j] = cross_in_tray([X_zoom[i, j], Y_zoom[i, j]])

# Créer la figure avec deux sous-graphiques
fig = plt.figure(figsize=(14, 7))

# Premier sous-graphe : vue d'ensemble
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)

# Ajouter les minima globaux
for (x_min, y_min) in minima:
    ax1.scatter(x_min, y_min, z_min, color='r', s=100)  # point rouge pour les minima globaux
    ax1.text(x_min, y_min, z_min, f'({x_min}, {y_min})', color='red')

# Ajouter des labels et un titre
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
ax1.set_title('Fonction Cross-in-Tray (Vue d\'ensemble)')

# Deuxième sous-graphe : vue restreinte
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(X_zoom, Y_zoom, Z_zoom, cmap='viridis', alpha=0.8)

# Ajouter les minima globaux
for (x_min, y_min) in minima:
    if -2 <= x_min <= 2 and -2 <= y_min <= 2:
        ax2.scatter(x_min, y_min, z_min, color='r', s=100)  # point rouge pour les minima globaux
        ax2.text(x_min, y_min, z_min, f'({x_min}, {y_min})', color='red')

# Ajouter des labels et un titre
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')
ax2.set_title('Fonction Cross-in-Tray (Zoom sur les minima)')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def schaffer_n2(xx):
    x = xx[0]
    y = xx[1]
    num = np.sin(x**2 - y**2)**2 - 0.5
    denom = (1 + 0.001 * (x**2 + y**2))**2
    return 0.5 + num / denom

# Créer une grille de points (x, y) pour la vue d'ensemble
x = np.linspace(-100, 100, 400)
y = np.linspace(-100, 100, 400)
X, Y = np.meshgrid(x, y)
Z = np.zeros_like(X)

# Calculer la valeur de la fonction pour chaque point de la grille
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        Z[i, j] = schaffer_n2([X[i, j], Y[i, j]])

# Coordonnées du minimum global
x_min = 0
y_min = 0
z_min = schaffer_n2([x_min, y_min])

# Créer une grille de points (x, y) pour la vue restreinte
x_zoom = np.linspace(-2, 2, 400)
y_zoom = np.linspace(-2, 2, 400)
X_zoom, Y_zoom = np.meshgrid(x_zoom, y_zoom)
Z_zoom = np.zeros_like(X_zoom)

# Calculer la valeur de la fonction pour chaque point de la grille restreinte
for i in range(X_zoom.shape[0]):
    for j in range(X_zoom.shape[1]):
        Z_zoom[i, j] = schaffer_n2([X_zoom[i, j], Y_zoom[i, j]])

# Créer la figure avec deux sous-graphiques
fig = plt.figure(figsize=(14, 7))

# Premier sous-graphe : vue d'ensemble
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='coolwarm', alpha=0.8)

# Ajouter le minimum global
ax1.scatter(x_min, y_min, z_min, color='r', s=100)  # point rouge pour le minimum global
ax1.text(x_min, y_min, z_min, 'Minimum Global', color='red')

# Ajouter des labels et un titre
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
ax1.set_title('Fonction Schaffer N.2 (Vue d\'ensemble)')

# Deuxième sous-graphe : vue restreinte
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(X_zoom, Y_zoom, Z_zoom, cmap='plasma', alpha=0.8)

# Ajouter le minimum global
ax2.scatter(x_min, y_min, z_min, color='r', s=100)  # point rouge pour le minimum global
ax2.text(x_min, y_min, z_min, 'Minimum Global', color='red')

# Ajouter des labels et un titre
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')
ax2.set_title('Fonction Schaffer N.2 (Zoom sur le minimum)')

plt.show()
```
